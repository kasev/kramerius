For this task, we will explore the possibilities of https://www.digitalniknihovna.cz and kramerius api.

The base url of a query in digitalniknihovna looks like this:

`https://www.digitalniknihovna.cz/mzk/search?q=`

A complete query can then look like this:

`https://www.digitalniknihovna.cz/mzk/search?q=demokracie&doctypes=periodical&collections=vc:cbed6e45-c63e-431f-8b16-8ff2c28e8a6f&from=1980&to=2000&sort=earliest`

The parameters:
* "**q**" (the query term itself): "demokracie"
* "**doctypes**" (Document Type): "periodical" (Newspaper)
* "**collections**" : vc:cbed6e45-c63e-431f-8b16-8ff2c28e8a6f (code for "Daily Press")
* Publishing Year:
    * **from**
    * **to**
* "sort" : sorting of the output

Query without query term is also functional:

`https://www.digitalniknihovna.cz/mzk/search?doctypes=periodical&collections=vc:cbed6e45-c63e-431f-8b16-8ff2c28e8a6f&from=1980&to=2000&sort=earliest`

```python
%%time
df = pd.DataFrame()
step=1000
for start in range(0, 100000, step):
    url = "https://kramerius.mzk.cz/search/api/v5.0/search?q=(fedora.model:page)%20AND%20(datum_begin:%5B*%20TO%202000%5D%20AND%20datum_begin:%5B1980%20TO%20*%5D)&q1=demokracie&wt=json&rows={0}&start={1}".format(step, start)
    resp = requests.get(url)
    resp_json = resp.json()
    df = df.append(pd.DataFrame(resp_json["response"]["docs"]))
```

# Get OCR full-text

```python
"https://kramerius.mzk.cz/search/api/v5.0/item/{0}/streams/TEXT_OCR"
```

```python
for pid in df[df["dostupnost"] != "private"]["PID"]:
    url = "https://kramerius.mzk.cz/search/api/v5.0/item/{0}/streams/TEXT_OCR".format(pid)
    resp = requests.get(url)
    print(resp.content.decode("utf-8"))
```

```python
url = "https://kramerius.mzk.cz/search/api/v5.0/item/uuid:3c712730-0a1d-11eb-b34d-005056827e51"
requests.get(url).content
```

```python
def extract_ocr(row):
    if row["dostupnost"] == "public":
        try:
            url = "https://kramerius.mzk.cz/search/api/v5.0/item/{0}/streams/TEXT_OCR".format(row["PID"])
            resp = requests.get(url)
            fulltext = resp.content.decode("utf-8")
        except:
            fulltext = ""
        return fulltext
    else:
        return ""
df["fulltext"] = df.apply(lambda row: extract_ocr(row), axis=1)
```

```python
fulltexts = " ".join(df["fulltext"].tolist())
```
